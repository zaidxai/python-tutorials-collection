## Important Functions of pandas

```python
import pandas as pd
import numpy as np
```

**Series:**

```python
ser=pd.Series(np.random.rand(15))
print(ser)
print(type(ser))
```

**DataFrame:**

```python
df=pd.DataFrame(np.random.rand(334,5),index=np.arange(334))
print(df)
print(type(df))
```

```python
print(df.describe())
print(df.dtypes)
```

**Changing Values**

```python
df.loc[0][0]='Zaid'
print(df.head(6))
print(df.dtypes)
```

**Sorting Values**

```python
print(df.index)
print(df.columns)
print(df.sort_index(axis=0,ascending=True))
# df.loc[0][0]=654
```

**Setting Value**

```python
df.loc[0,0]=654
print(df.head(3))
df.columns=list("ABCDE")
print(df.head(2))
df.loc[0,'A']=6540
print(df.head(3))
print(df.loc[[1,2],['C','D']])
```

**Acessing Rows**

```python
print(df.loc[:,['C','D']])
```

**Acessing Columns**

```python
print(df.loc[[1,2],:])
```

**Complex Quesries**

```python
print(df.loc[(df['B']<0.5) & (df['C']<0.5)])
```

**If we are counting from indexes**

```python
print(df.iloc[0,4])
```

```python
# print(df.drop(['A'],axis=1))
# print(df.info())
df.rename(columns={'A':'a'},inplace=True)
print(df.head(5))
```

**Dropping Columns**

```python
df['B']=df['B'].astype(int)
print(df.info())
print(df.head(5))
```
