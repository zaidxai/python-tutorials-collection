# Data Selection in Series

## Series as one-dimensional array

```python
import pandas as pd
data = pd.Series([0.25, 0.5, 0.75, 1.0],
index=['a', 'b', 'c', 'd'])
data['e'] = 1.25
data
```

```python
#slicing by explicit index
print(data['a':'c'])
#slicing by implicit index
print(data[0:2])
#masking
print(data[(data>0.3) & (data<0.8)])
#fancy indexing
print(data[['a','e']])
```

## Indexers: loc, iloc, and ix

```python
data = pd.Series(['a', 'b', 'c'], index=[1, 3, 5])
print(data)
# explicit index when indexing
print(data[1])
# implicit index when indexing
print(data[1:3])
```

**Loc(Explicit Index)**

```python
#loc attribute always refrence to the explicit index
print(data.loc[1])
print(data.loc[1:3])
```

**iLoc(Implicit Index)**

```python
#loc attribute always refrence to the implicit index
print(data.iloc[1])
print(data.iloc[1:3])
```

# Data Selection in DataFrame

**DataFrame as a dictionary**

```python
area = pd.Series({'California': 423967, 'Texas': 695662,'New York': 141297, 'Florida': 170312,'Illinois': 149995})
pop = pd.Series({'California': 38332521, 'Texas': 26448193,'New York': 19651127, 'Florida': 19552860,'Illinois': 12882135})
data = pd.DataFrame({'area':area, 'pop':pop})
data
```

```python
#acceessing via dictionary style indexing
print(data['area'])
print(data.area)
```

```python
data['density']=data['pop']/data['area']
data
```

**DataFrame as two-dimensional array**

```python
data.values
```

```python
#Transpose
data.T
```

```python
print(data.values[0])
print(data['area'])
```

**Implicit Indexing**

```python
print(data)
print(data.iloc[:3,1:3])
```

**Explicit Indexing**

```python
print(data.loc[:'Illinois', :'pop'])
```

**Loc with masking and fency indexing**

```python
data.loc[data.density>100,['pop','density']]
```

```python
#modification of values
data.iloc[0, 2] = 90
data
```

```python
#indexing refers to columns, slicing refers to rows:
print(data['Florida':'Illinois'])
print(data[1:3])
print(data[data.density > 100])
```

## Specifying Set Arithmetic for Joins

```python
df1 = pd.DataFrame({'name': ['Peter', 'Paul', 'Mary'],
                    'food': ['fish', 'beans', 'bread']},
                    columns=['name', 'food'])
df2 = pd.DataFrame({'name': ['Mary', 'Joseph'],
                    'drink': ['wine', 'beer']},
                    columns=['name', 'drink'])
print(pd.merge(df1,df2))
# intersection of the two sets of inputs
#pd.merge(df6, df7, how='inner')
```

```python
#union of the input columns
print(pd.merge(df1, df2, how='outer'))
```

## Example: US States Data

```python
pop = pd.read_csv('state-population.csv')
areas = pd.read_csv('state-areas.csv')
abbrevs = pd.read_csv('state-abbrevs.csv')
```

```python
print(pop.head())
print(areas.head())
print(abbrevs.head())
```

```python
merged = pd.merge(pop, abbrevs, how='outer',
                  left_on='state/region', right_on='abbreviation')
merged
```

```python
merged = merged.drop('abbreviation', axis=1)
merged
```

```python
merged[merged['population'].isnull()]
```

```python
merged.loc[merged['state/region'] == 'PR', 'state'] = 'Puerto Rico'
merged.loc[merged['state/region'] == 'USA', 'state'] = 'United States'
merged.isnull().any()
```
