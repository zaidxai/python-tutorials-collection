**Adding Columns**

```python
df['Total']=df['HP']+df['Attack']+df['Sp. Atk']+df['Sp. Def']+df['Speed']
df
```

**Filtering and Saving file**

```python
new_df=df.loc[(df['Type 1']=='Grass') & (df['Type 2']=='Poison') & (df['HP']>70)]
```

```python
new_df
```

```python
new_df.to_csv('filtered.csv',index=False)
```
